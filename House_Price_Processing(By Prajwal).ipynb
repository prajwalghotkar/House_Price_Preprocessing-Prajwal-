##### Fix the Outlier (Capping and Flooring) - Column n_hot_rooms

##### Print 99 Percentile Value column n_hot_rooms

```python
print(np.percentile(df1.n_hot_rooms,[99]))
```

##### Store 99 Percentile Value to a variable as upper limit.

```python
upper_limit=np.percentile(df1.n_hot_rooms,[99])[0]
print(upper_limit)
```

##### Display all records greater than the upper limit 

```python
display(df1[(df1.n_hot_rooms>upper_limit)])
```

##### Update upper limit value is greate than 3* upper limit update with 3* upper limit

```python
df1.n_hot_rooms[(df1.n_hot_rooms>3*upper_limit)]=3*upper_limit
display(df1[(df1.n_hot_rooms>upper_limit)])
```

##### Fix the Outlier (Capping and Flooring) - Column rainfall

###### Identify 1 Percentile Value

```python
print(np.percentile(df1.rainfall,[1]))
```

##### Store 1 Percentile Value to Variable

```python
lower_limit=np.percentile(df1.rainfall,[1])[0]
print(lower_limit)
```

##### Disaply Records less than lower limit

```python
display(df1[(df1.rainfall<lower_limit)])
```

##### If Value less than 0.3 * lower limit update with 0.3 * lower_limit

```python
df1.rainfall[(df1.rainfall < 0.3 * lower_limit)] = 0.3 * lower_limit
display(df1[(df1.rainfall < lower_limit)])
```
